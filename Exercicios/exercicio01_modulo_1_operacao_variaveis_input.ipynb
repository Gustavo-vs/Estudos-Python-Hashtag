## Parte 1 - Operações e Variáveis
Crie um progrma que imprima (print) os principais indicadores da loja Hashtag&Drink no último ano <br>
obs: faça tudo usando variáveis.

Valores do último ano:

Quantidade de vendas de Coca = 150<br>
Quantidade de vendas de Pepse = 130<br>
Preço unitário da Coca = 1,50<br>
Preço unitáro da Pepsi = 1,50<br>
Custo da loja: 2.500,00


Use o bloco abaixo paa criar todas as variáveis que precisar.

```python
qtde_vendas_coca = 150
qtde_vendas_pepsi = 130
preco_coca = 1.50
preco_pepsi = 1.50
custo_loja = 2500.00
```

 1. Qual foi o faturamento de Pepsi da loja?

```python
faturamento_pepsi = preco_pepsi * qtde_vendas_pepsi
print(f"Este foi o faturamento em cima do refrigerante Pepsi: R${faturamento_pepsi}")
```

2. Qual foi o Faturamento de Coca da Loja?

```python
faturamento_coca = preco_coca * qtde_vendas_coca
print(f"Este foi o faturamento em cima do refrigerante Coca: R${faturamento_coca}")
```

3. Qual foi o Lucro da loja?

```python
lucro_loja = faturamento_coca + faturamento_pepsi - custo_loja
print(f"O lucro da loja Hashtag&Drink foi de R${lucro_loja}")
```

4. Qual foi a Margem de lucro da loja? (Lembre-se, margem = lucro/faturamento). Não precisa formatar em percentual

```python
faturamento_total = faturamento_pepsi + faturamento_coca
margem_loja = lucro_loja / faturamento_total

print(f"A Margem de lucro da loja foi R${margem_loja}")
```

Nota: neste primeiro exercico tive um erro comum onde tentei concatenar (juntar) uma str(string) com valores do tipo int (inteiro) ou <br>
do tipo float(números decimais) e isso resultou em TypeError. Isso aconteceu porque o python não lê a concatenação de str <br>
com valores (int/float) quando mandamos imprimir algo, por este motivo devemos deixar tudo oque<br>
 estiver dentro do print() ou como valores ou como str, como nestes exercicios eu preferi colocar um texto<br>
dentro do print() precisamos mostrar ao python que tudo dentro do print é string. E como fazemos isso?<br>

Temos varias maneiras de fazer isso sendo elas:

#### 1. Usando f-strings
As f-strings permitem incluir variáveis diretamente dentro da string, convertendo-as automaticamente. 

idade = 25<br>
preco = 19.<br>
print(f"Idade: {idade}, Preço: {preco}")<br>

Saída: Idade: 25, Preço: 19.99<br>
OBS: Esta foi a forma que utilizamos pois assim conseguimos colocar a variavel dentro do texto e não como formato de concatenação

#### 2. Usando a função str() com +
Converte explicitamente o número para string antes da concatenação. 

nome = "Produto"<br>
valor = 50<br>
print(nome + " - R$ " + str(valor))<br>

Saída: Produto - R$ 50<br>
Desta forma usamos a concatenção porém transformamos a variável "valor" em str antes da concatenação

#### 3. Usando .format()

item = "Maçã"<br>
preco = 2.50<br>
print("O preço de {} é {}.".format(item, preco))<br>

Saída: O preço de Maçã é 2.5.<br>
Nesta caso usamos os {} para representar as variáveis que serão utilizdas e em .format colocamos o nome das variáveis que desejamos incluir no texto no lugar do {}

#### 4. Usando vírgula no print()
A vírgula adiciona automaticamente um espaço entre os elementos sem precisar converter para string.

item = "Maçã"<br>
preco = 2.50<br>
print("Preço:", preco)<br>

Saída: Preço: 2.5<br>
Está é uma forma mais fácil de se fazer porém a variável ficará para fora das ""

Cuidado Especial

Nunca use o print com =<br>
Um erro comum é fazer algo como:

print = algum_valor

. Isso não pode ser feito em momento algum. Isso porque colocar o sinal de = atribui um valor ao print, que é um comando interno do Python<br>
· Fazer isso vai mudar o que a função print faz e vai gerar um erro nos códigos seguintes. Vamos ver na prática.<br>
· Caso aconteça sem querer, reinicie o Kernel para corrigir
