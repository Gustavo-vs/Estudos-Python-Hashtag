# 1. Validador de Empréstimo (Matemática + Lógica Pro)<br>
    
   •Inputs: valor_emprestimo, salario_cliente e tempo_casa_anos.<br>
   •Validação: Verifique se todos os campos foram preenchidos e se são numéricos (.isnumeric()).<br>
    
   •Regras:<br>
   •Calcule a parcela: valor_emprestimo / 36.<br>
   •Regra A (2 anos ou mais de casa): A parcela não pode ultrapassar 30% do salário.<br>
   •Regra B (Menos de 2 anos de casa): A parcela não pode ultrapassar 20% do salário.<br>
   •Saída: Exiba se o empréstimo foi aprovado ou negado e o valor da parcela.<br>

```python
valor_emprestimo = input("Qual é o valor do Empretimo: ")
salario_cliente = input("Informe seu salário: ")
tempo_casa_anos = input("Á quantos anos você trabalha em sua empresa atual")

if valor_emprestimo.isnumeric() and salario_cliente.isnumeric() and tempo_casa_anos.isnumeric():
    valor_emprestimo = float(valor_emprestimo)
    salario_cliente = float(salario_cliente)
    tempo_casa_anos = int(tempo_casa_anos)

    valor_parcela = valor_emprestimo / 36

    if tempo_casa_anos >= 2:
        valor_maximo_parcela = salario_cliente * 0.3
        if valor_parcela > valor_maximo_parcela:
            print(f"Infelismente seu empréstimo foi negado pois o valor da parcela ultrapassa os 30%. Sua parcela é de {valor_parcela:.2f}")
        else:
            print(f"Parabéns seu empréstimo foi aprovado ele é de R${valor_emprestimo:.2f} com parcelas de {valor_parcela:.2f}")
    else:
        valor_maximo_parcela = salario_cliente * 0.2
        if valor_parcela > valor_maximo_parcela:
            print(f"Infelismente seu empréstimo foi negado pois o valor da parcela ultrapassa os 30%. Sua parcela é de {valor_parcela:.2f}")
        else:
            print(f"Parabéns seu empréstimo foi aprovado ele é de R${valor_emprestimo:.2f} com parcelas de {valor_parcela:.2f}")
else:
    print("Preencha todos os campos com as informações válidas")
```

# 2. Sistema de Login (Segurança e Strings Pro)

Inputs: usuario_digitado e senha_digitada.<br>

Regras:

#   💰 Exercício 3: CRM de Vendas (Bônus e Meta Global)<br>
Cenário: Você está desenvolvendo o módulo de premiação de um sistema de vendas. O objetivo é calcular quanto um vendedor receberá de bônus baseado no desempenho do mês.

### O que o seu código deve fazer:<br>

1. Inputs do Usuário:<br>
•nome_vendedor<br>
•valor_venda

2. Tratamento e Validação (O "Padrão Pro"):<br>
•Use .strip() e .replace(" ", "") no valor da venda para garantir que espaços não quebrem o cálculo.<br>
•Verifique se os campos foram preenchidos.<br>
•Verifique se o valor da venda é numérico (.isnumeric()).

3. Regra de Negócio 1: A Escala de Porcentagem<br>
•Se venda < 10000: Bônus de 0%.<br>
•Se venda entre 10.000 e 30.000: Bônus de 5% sobre o valor vendido.<br>
•Se venda > 30.000: Bônus de 10% sobre o valor vendido.

4. Regra de Negócio 2: O Prêmio Superação (Meta Global)<br>
•Independente da porcentagem anterior, se o vendedor sozinho atingir a meta global da loja de R$ 50.000, ele ganha um adicional fixo de R$ 1.500,00.

5. Saída de Dados:<br>
•Exiba o bônus final total que o vendedor receberá.

```python
#Constante
meta_global_loja = 50000

#inputs usuario
nome_vendedor = input("Digite seu Nome e Sobrenome: ").upper()
valor_venda = input("Insira o valor da venda: ").strip() #.strip() retira o espaço das ponta seja no iniciou ou fim da digitação

if nome_vendedor and valor_venda:
    if valor_venda.isnumeric():
        valor_venda = float(valor_venda)
        if valor_venda >= meta_global_loja:
            print("Você Ganhou um adicional de R$1500,00")
        else:   
            if 10000 < valor_venda < 30000:
                percent_bonus = 0.05
            
            elif valor_venda > 30000:
                percent_bonus = 0.1
            
            else:
                percent_bonus = 0
            
            bonus_venda = valor_venda * percent_bonus

            if 10000 < valor_venda < 30000:
                print(f"Seu bonus foi de: R${bonus_venda}")
            elif valor_venda > 30000:
                print(f"Seu bonus foi de: R${bonus_venda}")
            else:
                print("Você não teve bonus este mês pois suas vendas foram abaixo de R$10000,00")
            
    else:
        print("Digite um valor válido no campo do Valor de venda!")
else:
    print("Digite os campos corretamente")


```

## Refatoração Pro Gemini

```python
# Constante
meta_global_loja = 50000

# Inputs
nome_vendedor = input("Digite seu Nome e Sobrenome: ").upper().strip()
valor_venda = input("Insira o valor da venda: ").strip().replace(" ", "")

if nome_vendedor and valor_venda.isnumeric():
    valor_venda = float(valor_venda)
    bonus_total = 0
    
    # 1. Cálculo da Porcentagem (Escala)
    if valor_venda > 30000:
        bonus_total = valor_venda * 0.10
    elif valor_venda >= 10000:
        bonus_total = valor_venda * 0.05
    # Se for menor que 10k, o bonus_total continua 0 (não precisa de else: pass)

    # 2. Verificação da Meta Global (Prêmio Extra Independente)
    # Usamos IF e não ELIF, pois queremos que o Python teste isso SEMPRE
    if valor_venda >= meta_global_loja:
        bonus_total += 1500  # Isso é o mesmo que: bonus_total = bonus_total + 1500
        print("🎉 Parabéns! Você atingiu a Meta Global e ganhou +R$ 1.500,00 fixos!")

    # 3. Saída Única (Evitando repetição de prints)
    if bonus_total > 0:
        print(f"Olá {nome_vendedor}, seu bônus total este mês é de: R${bonus_total:.2f}")
    else:
        print(f"Olá {nome_vendedor}, infelizmente você não atingiu o bônus mínimo este mês.")

else:
    print("Erro: Verifique se preencheu o nome e se inseriu apenas números nas vendas.")
```

# 🏦 Exercício 4: Classificador de Investimentos<br>
Cenário: Uma corretora de valores precisa classificar seus clientes em três categorias automáticas baseadas no saldo que eles possuem em conta para oferecer produtos específicos.

O que o seu código deve fazer:

1. Inputs do Usuário:<br>
•nome_cliente<br>
•saldo_conta

2. Tratamento e Validação:<br>
•Remova espaços extras e trate os números (use .replace(" ", "") e .strip()).<br>
•Verifique se o saldo é numérico antes de converter para float.

3. Regras de Classificação (Lógica de Refatoração):<br>
•Standard: Saldo abaixo de R$ 1.000,00.<br>
•Prata: Saldo entre R$ 1.000,00 e R$ 10.000,00.<br>
•VIP: Saldo acima de R$ 10.000,00.

```python
nome_cliente = input("Digite o nome do cliente aqui: ").strip().replace(" ", "")
saldo_conta = input("Insira o valor da conta: ").strip().replace(" ", "")

if nome_cliente and saldo_conta:
    if saldo_conta.isnumeric():
        saldo_conta = float(saldo_conta)
        categoria = "Standard"
        if 1000 <= saldo_conta < 10000:
            categoria = "Prata"
        elif saldo_conta >= 10000:
            categoria = "VIP"
       
        print(f"O seu investimento é do nivel {categoria}")
        
    else:
        print("Insira um valor válido no campo de saldo cliente!!")
else:
    print("Preencha todos os campos solicitados!!")
```
