# 🛡️ Desafio Boss: Checkout de E-commerce Inteligente<br>
Cenário: Você foi contratado para criar o motor de cálculo de um carrinho de compras. O sistema deve receber o valor total, aplicar descontos baseados no tipo de cliente e calcular o frete baseado na região.

## 1. Entradas (Inputs):<br>
 • nome_cliente<br>
 • valor_compra<br>
 • cupom_desconto (O usuário pode digitar qualquer coisa aqui).<br>
 • regiao_entrega (Ex: "Norte", "Sul", "Sudeste", etc.).

## 2. Regras de Negócio (A Lógica):<br>
Validação Pro: * O valor da compra deve ser numérico e maior que zero.<br>
 • O nome do cliente não pode estar vazio.<br>
 • Use .strip().lower().replace(" ", "") onde for necessário para evitar erros de digitação.<br>
 
 Sistema de Cupom:<br>
 • Se o cupom for "BEMVINDO10": 10% de desconto.<br>
 • Se o cupom for "MESTREPYTHON": 20% de desconto.<br>
 • Se for qualquer outro ou vazio: 0% de desconto.<br>
  
 Regra do Frete (Após o Desconto):<br>
 • Região "sudeste": Frete Fixo R$ 15,00.<br>
 • Região "sul" ou "centro-oeste": Frete Fixo R$ 25,00.<br>
 • Outras regiões: Frete Fixo R$ 40,00.<br>
  
 Frete Grátis: Se o valor final da compra (já com desconto) for maior ou igual a R$ 250,00, o frete deve ser R$ 0,00, independente da região.<br>
 • Brinde Especial (O detalhe do Boss):<br>
 • Se o cliente for da região "sudeste" E o valor final (com desconto) for maior que R$ 500,00, ele ganha um brinde: "Fone de Ouvido Premium".<br>

## 3. Requisitos de Refatoração (O que eu vou avaliar):<br>
 • Apenas UM print final resumindo tudo (Nome, Valor Inicial, Desconto Aplicado, Frete e Valor Total a Pagar).<br>
 • Use variáveis auxiliares para porcentagem_desconto, valor_frete, brinde e valor_final.<br>
 • Comece com brinde = "Nenhum".

```python
nome_cliente = input("Digite o nome e sobrenome do cliente: ").upper().strip()
valor_compra = input("Insira o valor da compra: ").strip().replace(",", ".")#Substitui , por . para que não de erro na hora de transformar em float
cupom_desconto = input("Insira seu cupom de desconto: ").upper().strip().replace(" ", "")
regiao_entrega = input("Indique sua orientação geográfica (N, S, L, O, NE, SE, NO, SO): ").upper().strip().replace(" ", "")


if nome_cliente and valor_compra and regiao_entrega:

#Retira o . da variavel e junta os numeros ela faz isso somente uma vez dentro da variavel para que isso possa ser lido pelo .isnumeric() como um valor numerico
    if valor_compra.replace(".", "", 1).isnumeric():
        valor_compra = float(valor_compra)
        if valor_compra > 0:
            
            percent_cupom = 0
            #valor da variavel só ira mudar caso uma das seguintes condições forem verdadeiras
            if cupom_desconto == "BEMVINDO10":
                percent_cupom = 0.1
            elif cupom_desconto == "MESTREPYTHON":
                percent_cupom = 0.2
            
            #Valor da compra com o desconto aplicado
            valor_compra_desconto = valor_compra - (valor_compra * percent_cupom)
            
            frete_regiao = 40.00
            #valor da variavel só ira mudar caso uma das seguintes condições forem verdadeiras
            if valor_compra_desconto >= 250.00:
                frete_regiao = 0.00
            else:
                if regiao_entrega == "SE":
                    frete_regiao = 15.00
                elif regiao_entrega == "S" or regiao_entrega == "SO":
                    frete_regiao = 25.00
            
            brinde = "NENHUM"

            if regiao_entrega == "SE" and valor_compra_desconto > 500.00:
                brinde = "Fone de Ouvido Premium como brinde"
                print("Parabés você ganhou um {} \n".format(brinde))
            
            valor_final = valor_compra_desconto + frete_regiao
            valor_desconto = valor_compra * percent_cupom
            print(f"Olá {nome_cliente} o valor da sua compra foi de R${valor_compra:.2f} e o seu desconto foi de R${valor_desconto:.2f}, \no valor do frete foi de R${frete_regiao:.2f} e o valor total a pagar é R${valor_final:.2f} e seu birnde é {brinde}")

        else:
            print("Valor da compra incompativel por favor insira um valor válido")
    else:
        print("Insira um valor numérico no campo valor da compra!!")
else:
    print("Preencha todos os campos corretamente!!")

print("Teste Resolv1")
```
