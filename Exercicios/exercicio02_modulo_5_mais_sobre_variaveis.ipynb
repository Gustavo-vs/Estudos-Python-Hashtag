# Exercício 1: O Analista de Salário<br>
Este foca na base: entrada de dados, conversão de tipos e formatação de texto.

• O Desafio: Crie um programa que peça o nome do usuário, o cargo dele e o salário bruto mensal.

Requisitos:

1. Converta o salário (input) para um número decimal (float).

2. Calcule o bônus anual (digamos, 15% do salário mensal).

3. Imprima uma frase usando f-string que diga: "Olá [NOME], seu cargo é [CARGO] e seu bônus é de R$ [VALOR]".

4. Ao final, use a função type() para mostrar na tela qual é o tipo da variável que armazena o salário após a conversão.

```python
nome = input("Insira seu nome completo: ")
cargo = input("Insira seu cargo: ")
salario_bruto = float(input("Insira seu salário bruto: "))
bonus_anual = salario_bruto * 0.15 # 0.15 é equivalente á 15% lembre-se sempre em contas de porcentagem tranformar o valor

print(f"Olá {nome} seu cargo dentro de nossa empresa é {cargo} e você recebe de bônus R${bonus_anual:.2f}") 
# utilizamos :.2f em bonus_anual isso significa que estou pedindo para o print() me mostrar somento 2 casas após a virgula
print(type(bonus_anual))
```

# Exercício 2: O Manipulador de Código de Produto<br>
Este foca em operações com strings e matemática básica.

• O Desafio: Peça ao usuário para digitar o código de um produto no formato:<br> 
NOMEDOPRODUTO-ANOFABRICACAO (Ex: TECLADO-2024). Além disso, peça o preço de custo <br>
e a margem de lucro desejada (em %).

Requisitos:

1. Use operações de string para deixar o nome do produto todo em maiúsculas.
2. Calcule o preço de venda: $custo + (custo \times \frac{lucro}{100})$.
3. Mostre no print o tipo da variável do preço de venda.
4. Imprima: "O produto [PRODUTO] fabricado em [ANO] será vendido por R$ [PRECO_VENDA]".

```python
# Recebe o código e já trata para maiúsculas por conta do (.upper()) (padrão de banco de dados)
codigo_produto = input("Digite o nome do produto e o ano de fabricação (PRODUTO-ANOFABRICAÇÃO): ").upper()
preco_custo = float(input("Digite o preço de custo do produto: "))
margem_lucro = float(input("Digite sua margem de lucro em % : "))

# Divide a string pelo hífen para separar Nome do Produto e Ano
nome_produto = codigo_produto.split("-") 

margem_lucro_percent = margem_lucro / 100
preco_venda = preco_custo + (preco_custo * margem_lucro_percent )

# O .split("-") gera uma lista: ['NOME', 'ANO']
# nome_produto[0] acessa o primeiro item da lista (o nome do produto)
# nome_produto[1] acessa o segundo item da lista (o ano de fabricação)
print("O produto {} fabricado em {} será vendido por R${:.2f}".format(nome_produto[0], nome_produto[1], preco_venda))
print(type(preco_venda))






```

# Exercício 3: O Simulador de Investimento Simples<br>
Misturando tudo com uma lógica mais estruturada.

• O Desafio: Peça o nome de um banco, o valor de um investimento inicial e a taxa de juros anual (em %).

Requisitos:

1. Garanta que o valor do investimento seja um número inteiro (int) e a taxa seja decimal (float).

2. Calcule quanto o usuário terá após 1 ano (Investimento + Juros).

3. Realize uma operação de string: Repita o nome do banco 3 vezes com espaços entre eles (Ex: "Nubank Nubank Nubank") e salve em uma variável.

4. Exiba uma mensagem completa: "No banco [BANCOS], seu saldo final será de [VALOR_FINAL]".

5. Mostre o tipo da variável que armazena a taxa de juros.

```python
nome_banco = input("Insira o nome do seu banco: ").upper()
valor_investimento = int(input("Insira o valor que deseja investir: "))
valor_taxa_anual = float(input("Insira o valor da taxa de juros anual que você deseja: "))

retorno_investimento_1ano = (valor_investimento * (1 + valor_taxa_anual / 100))

nome_repetido_banco = (nome_banco + " ") * 3

print(f"No banco {nome_repetido_banco}, seu saldo final será de {retorno_investimento_1ano:.2f}" )
print(type(valor_taxa_anual))
```
