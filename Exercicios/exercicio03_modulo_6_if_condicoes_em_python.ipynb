```python
vendas_funcionario1 = 1000
vendas_funcionario2 = 770
vendas_funcionario3 = 2700
meta_vendas = 1000

if vendas_funcionario1 >= meta_vendas:
    bonus_funcionario1 = vendas_funcionario1 * 0.1
    print("O funcionário 1 ganhou {} de bônus ".format(bonus_funcionario1))
else:
    print("O funcionário 1 não ganhou bônus")

if vendas_funcionario2 >= meta_vendas:
    bonus_funcionario2 = vendas_funcionario2 * 0.1
    print("O funcionário 2 ganhou {} de bônus ".format(bonus_funcionario2))
else:
    print("O funcionário 2 não ganhou bônus")

if vendas_funcionario3 >= meta_vendas:
    bonus_funcionario3 = vendas_funcionario3 * 0.1
    print("O funcionário 3 ganhou {} de bônus ".format(bonus_funcionario3))
else:
    print("O funcionário 3 não ganhou bônus")

#crie seu código aqui
```

# Terceiro Código (Refatorado Pro com ajuda gemini)

# 💡 Como elevar o nível (A Refatoração "Pro")
Em vez de usar o pass (que verifica se está TUDO BEM para não fazer nada), o ideal é inverter a lógica: Verificar apenas se o estoque está BAIXO. Veja como o código fica muito mais limpo e fácil de ler (Menos é Mais):

```python
nome_produto = input("Nome do produto: ").upper()
categoria_produto = input("Categoria: ").upper()
qtde_estoque_produto = input("Quantidade: ")

if nome_produto and categoria_produto and qtde_estoque_produto.isnumeric():
    qtde_estoque_produto = int(qtde_estoque_produto)
    
    # Definimos o estoque mínimo de acordo com a categoria
    if categoria_produto == "ALIMENTOS":
        estoque_min = 50
    elif categoria_produto == "BEBIDAS":
        estoque_min = 75
    elif categoria_produto == "LIMPEZA":
        estoque_min = 30
    else:
        estoque_min = 0 # Categoria não cadastrada
        print("Categoria não reconhecida.")

    # Só fazemos UM print de solicitação no final se estiver abaixo do mínimo
    if qtde_estoque_produto < estoque_min:
        print(f"Solicitar {nome_produto} à equipe de compras, temos apenas {qtde_estoque_produto} em estoque.")
    
else:
    print("Preencha os campos corretamente com valores numéricos no estoque!")
```

# 🔍 Por que essa versão é melhor?

1.Sem Repetição: O print da solicitação de compra aparece apenas uma vez no código. Se você precisar mudar o texto da mensagem amanhã, muda em um lugar só.

2.Lógica Direta: Em vez de usar pass para quando o estoque está alto, nós simplesmente não fazemos nada. O programa só age se o estoque for menor que o mínimo.

3.Conversão Única: Fizemos qtde = int(qtde) uma única vez no início, evitando ter que escrever int() em todas as linhas.

O uso do pass é exatamente para isso! Em Python, você não pode deixar um bloco de controle (como um if ou uma função) totalmente vazio, senão o código gera um erro de indentação. O pass serve como um "espaço reservado" ou um "cala a boca" para o Python, dizendo: "Eu sei que aqui deveria ter algo, mas por enquanto não quero fazer nada".
