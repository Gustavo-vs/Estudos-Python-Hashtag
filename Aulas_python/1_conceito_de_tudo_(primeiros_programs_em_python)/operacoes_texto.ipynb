Operações com texto (string) (sem o uso de variaveis)

```python
print("meu nome é Gustavo")

print("meu nome é" + " Gustavo") #isso é uma concatenação de textos (strings) onde eu junto um tetxo com outro

print("G" in "Gustavo") 
# Funcionamento do comando in : 
# oque_eu_quero_procurar in onde_vou_procurar (ou melhor dizendo dentro_de_que_vou_procurar)
```

A operação in faz um comparação entre uma string e outra neste caso esta acontecendo o seguinte <br>
a letra "G" esta presente em "Gustavo" como a resposta é sim ele me retorna True caso não ele me retornatia False <br>
então na programação é como se subistituissemos a frase 'esta presente em' pelo comando/operação in e nuca se esqueça <br>
o comando/operação in é Case sensitive Oque isso significa ? Significa que ele liga se a letra ou palavra que você esta procurando <br>
esta em minusculo ou MAIUSCULO, se eu tentasse procurar a letra "g" ao invés de "G" <br>
o terminal iria me retornar False pois a palavra não possui g minusculo e sim G maiusculo

```python
print("Gustavo" in "Meu nome é Gustavo")
# Funcionamento do comando in : 
# oque_eu_quero_procurar in onde_vou_procurar (ou melhor dizendo dentro_de_que_vou_procurar)

#neste exemplo procurei uma palavra dentro de uma frase -
# e note que ambos começam com a letra maiuscula -
# caso um começasse com letra maiuscula e minha procura fosse com o inicio minusculo eu teria False com resultado em meu terminal
```
