# Mudança do Tipo de Variáveis

Tenha em mente que todo o dado inserido dentro de um input sempre será considerado como uma str<br>
Aqui veremos como mudar o tipo de dado de um input para um valor numérico como int ou float

### Exemplo:<br>
faturamento = input("Insira o faturamento total da loja ")<br>
custo = input("Insira o custo total da loja")<br>
lucro = faturamento - custo<br>
print(f"O lucro da loja foi {lucro}")

Note que se você execultar esse código ele dara o seguinte erro: <br>

(TypeError: unsupported operand type(s) for -: 'str' and 'str')<br>

Isso significa que você está tentando fazer uma operação mátematica com duas strings e isso lógicamente dara errado

Vejá abaixo as formas corretas de modificar os valores das váriaveis do input para que a operação não de erro

```python
#Usando int
faturamento = int(input("Insira o faturamento total da loja "))
custo = int(input("Insira o custo total da loja"))
lucro = faturamento - custo
print(f"O lucro da loja foi {lucro}")

#Usando float
faturamento = float(input("Insira o faturamento total da loja "))
custo = float(input("Insira o custo total da loja"))
lucro = faturamento - custo
print(f"O lucro da loja foi {lucro}")

##Veja que para evitar o erro é só colocar seu input dentro de um parenteses e defini-lo como int, float ou str do lado de fora desse novo parenteses
# int -> muda para inteiro
# float -> muda para numeros decimais (ponto flutuante)
# str -> muda para texto (string)
```
