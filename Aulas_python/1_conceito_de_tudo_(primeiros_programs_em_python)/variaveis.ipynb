Por quê o uso de variaveis é importante ? imagine que você tem um código com mais de 150 linhas<br>
e durante todo esse código você ao invés de criar uma variavel para um nome de uma pessoa e criar uma variavel<br>
para um determinado valor você simplesmente tivesse colocado esse nome e esse valor repetidas vezes durante seu código<br>
porém a pesso a que você se referia foi embora da sua empresa, e entrou outra pessoa no lugar e agora você precisa atualizar isso em seu código<br>
como você não criou variaveis tera que percorrer todo o seu código e mudar o valor e o nome da pessoa, imagine você fazendo isso em um código<br>
de 400 e poucas linhas você iria surtar, então a importancia da variavel é essa ela amarzena um valor e representa esse valor por<br>
todo o seu código e se um dia esse valor mudar você só tera que modificar o valor da variavel e não ira precisar percorrer o código inteiro.<br>
