```python
faturamento = 150
custo = 60
```

```python
lucro = faturamento - custo
print(lucro)
```

Note que ao escrever esse código e execultalo ele mostrara um erro dizendo que as variaveis escritas no segundo bloco de códigos não foram definidas, por quê isso acontecesse? Isso acontece porque quando temos dois blocos de código e um depende do outro para funcionar eles devem ser execultados em oredem, neste caso precisamos execultar o bloco 1 e depois o bloco 2 pois o bloco dois necessita que as variaveis do bloco um sejam definidas para o código funcionar.

E se eu modificar os valors do bloco um? novamente sempre que temos dois blocos de código e um depende do outro um sempre devera ser execultado antes do outro então vamos pensar neste exemplo se rodarmos o primeiro bloco e depois o segundo meu resultado no teminal será 90 pois estou subitraindo a variavel (faturamento) pela variavel (custo) onde faturamento é igual a 150 e custo igual a 60, vamos dizer que agr meu custo aumentou para 80 e eu fui e modifiquei isso em minha variavel (custo) mas não execultei o bloco onde essa variavel esta neste caso no bloco 1, e ai eu vou e execulto somente o bloco dois onde tenho a operção de subitração das minhas variaveis e o print desse resultado o meu terminal me mostrar o valor 90 novamento pois ele ira fazer a conta em cima dos valores da sua ultima execulção que foram (faturamento) igual a 150 e custo igual a 60 caso você queira a saida do novo resultado você era que execultar o primeiro bloco novamente para que o segundo bloco receba os novos valores das variaveis. caso queira que tudo isso seja feito automatico onde todo o seu código seja lido de cima para baixo de uma vez é só clicar em RUN ALL que ele ira execultar todos os blocos de código de uma unica vez
