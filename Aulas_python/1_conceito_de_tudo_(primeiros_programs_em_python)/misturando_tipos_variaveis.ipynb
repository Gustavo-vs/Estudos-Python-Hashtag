#### 1. Usando f-strings
As f-strings permitem incluir variáveis diretamente dentro da string, convertendo-as automaticamente. 

```python
faturamento = 6000
custo = 2600.50
lucro = faturamento - custo

#Com .format
print("O faturamento da loja foi de {}; O custo da loja foi de {}; O lucro total da loja foi de {}" .format(faturamento, custo, lucro))

#Com f-string
print(f"O faturamento da loja foi de {faturamento}; O custo da loja foi de {custo}; O lucro total da loja foi de {lucro}")
```

#### 2. Usando a função str() com +
Converte explicitamente o número para string antes da concatenação. <br>
Desta forma usamos a concatenção porém transformamos a variável "valor" em str antes da concatenação

```python
faturamento = 6000
custo = 2600.50
lucro = faturamento - custo
print("O faturamento da loja foi de " + str(faturamento))
print("O custo da loja foi " + str(custo))
print("O lucro total da laja foi de " + str(lucro))
```

#### 3. Usando .format()
Nesta caso usamos os {} para representar as variáveis que serão utilizdas e em .format colocamos o nome das variáveis que desejamos incluir no texto no lugar do {}

```python
faturamento = 6000
custo = 2600.50
lucro = faturamento - custo
print("O faturamento da loja foi de {}; O custo da loja foi de {}; O lucro total da loja foi de {}" .format(faturamento, custo, lucro))

#Lembre-se a ordem do seu .format() será a ordem em que os valores das variaveis serão mostrados no print()
#Neste caso colocamos .format(faturamento, custo, lucro) então no primeiro {V1} será faturamento, no segundo {V2} custo, e no terceiro {V3} lucro
# V1 = variavel 1 = faturamento
# V2 = variavel 2 = custo
# V3 = variavel 3 = lucro
```
