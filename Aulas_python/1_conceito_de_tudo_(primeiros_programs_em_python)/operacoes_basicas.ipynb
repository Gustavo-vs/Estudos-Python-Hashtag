Operações matemáticas básicas sem o uso de váriaveis (somente print)

```python
print( 1 + 1 ) #soma
print( 1 - 1 ) #subitração
print( 5 * 7 ) #multiplicação 
print( 20 / 4 ) #divisão
print( 7 % 3 ) #mod (mostra o resto de uma divisão)
```

```python
print(1 + 3 * 5 / 2) #nessa operação você pode ver que o python segue a ordem das operçãoes
print((2  + 3) * 7) #nessa operação você pode ver que o python respeita os parenteses e priorisa oque esta dentro do parenteses
```
