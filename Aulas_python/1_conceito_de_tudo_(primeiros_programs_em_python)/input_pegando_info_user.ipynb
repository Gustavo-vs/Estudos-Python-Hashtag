Pegando as informações de um usúario

```python
name = input("Qual é o seu primeiro nome?")
sobrenome = input("Qual o seu segundo nome?")

print(name + ' ' + sobrenome)

# Qaundo se usa o comando input o terminal ira exigir do usuario a entrada de um dado então isso pode ser usa para sabe o -
# nome, idade, altura, etc... De um deerminado usúario.
```
