Se o usuário digitar "mil" em vez de 1000, o seu código vai travar com um erro chamado ValueError. Isso acontece porque o Python não sabe como converter letras em números inteiros na linha do int(faturamento).

Por que isso acontece?
A sua validação atual (if faturamento and custo:) só checa se o campo está vazio ou não. Ela não checa o que tem dentro do campo. Se tiver a letra "A", o Python entende como "verdadeiro" (existe conteúdo), tenta converter para inteiro e... BUM, o programa fecha.

## 🛠️ Como um Dev Back-end resolve isso?
Existem duas formas principais de proteger seu código contra esse tipo de erro:

### 1. Usando o método .isnumeric()
Este método de strings verifica se todos os caracteres ali dentro são números de 0 a 9.

```python
faturamento = input('Qual foi o faturamento da loja nesse mês?')
custo = input('Qual foi o custo da loja nesse mês?')

# isso significa que estou verificando se existe algum valor dentro de faturamento e custo seja ele texto ou valor numérico, 
# caso não tenha nenhum valor la dentro ou seja ''(campo vazio) ele sai do if e vai para o else
# verifico também se os valores dentro de faturamento e custo são valores numéricos isso é possivel por conta do .isnumeric
#Este método de strings verifica se todos os caracteres ali dentro são números de 0 a 9.

if faturamento.isnumeric() and custo.isnumeric():
    lucro = int(faturamento) - int(custo)
    print(f"O lucro foi de {lucro}")
else:
    print("Erro: Por favor, digite apenas números positivos.")
```

## 2. Usando try / except (O jeito profissional)

Na Ciência da Computação, chamamos isso de Tratamento de Exceções. <br>
Você diz ao Python: "Tente fazer isso, se der erro, faça aquilo em vez de travar".

```python
faturamento = input('Qual foi o faturamento da loja nesse mês?')
custo = input('Qual foi o custo da loja nesse mês?')    

try: # Tente fazer isso
    lucro = int(faturamento) - int(custo)
    print(f"O lucro foi de {lucro}")

# se der erro

except ValueError: # faça isso em vez de travar".
    print("Erro: Os valores digitados não são números válidos.")
```
