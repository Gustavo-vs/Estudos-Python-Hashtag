# Comparações Contraintuitivas<br>
Existem algumas comparações no Python que não são tão intuitivas quando vemos pela primeira vez, mas que são muito usadas, principalmente por<br>
programadores mais experientes.

É bom sabermos alguns exemplos e buscar sempre entender o que aquela comparação está buscando verificar.

### Exemplo 1:

Digamos que você está construindo um sistema de controle de vendas e precisa de algumas informações para fazer o cálculo do resultado da loja no fim de
um mês.



```python
faturamento = input('Qual foi o faturamento da loja nesse mês?')
custo = input('Qual foi o custo da loja nesse mês?')

# Digamos que o usuario não inseriu nada no primeiro ou no segundo input isso acarretara em um erro ao executar o código
# A seguir vamos mostrar duas formas de evitar isso

# Primeira forma:
# Aqui voce usa o comparador or para dizer - se o faturamento for igual a '' (campo vazio) ou o custo for igual a '' printe a mensagem
if faturamento == '' or custo == '':
    print("Preencha o faturamento e o custo correamente")
else:
    lucro = int(faturamento) - int(custo)
    print("O lucro da loja foi de {} reais".format(lucro))

# Segunda forma:
# Aqui entramos na comparação contraintuitiva

# isso significa que estou verificando se existe algum valor dentro de faturamento e custo seja ele texto ou valor numérico, 
# caso não tenha nenhum valor la dentro ou seja ''(campo vazio) ele sai do if e vai para o else
if faturamento and custo: 
    lucro = int(faturamento) - int(custo)
    print("O lucro da loja foi de {} reais".format(lucro))
else:
    print("Preencha o faturamento e o custo correamente")

# Em Python, objetos têm valores booleanos intrínsecos. Strings vazias são tratadas como False. 
# Ao usar if faturamento and custo:, estamos verificando a existência de conteúdo antes de realizar operações de conversão (int), 
# evitando erros de execução (ValueError).
```
