# Estrutura de decisão $elif$

Em Python temos os comandos que usamos em conjunto: $if,  elif$ e $else$

$if$ (teste para decisão): <br>
comandos executados caso o teste seja verdadeiro<br>

$elif$ (outro teste para decisão):<br>
comandos executados caso esse teste seja verdadeiro<br>

$else$:<br>
comandos executados caso todos os testes anteriores sejam falsos

## 1. Para que serve?

Ele permite que seu programa tome mais de duas decisões.<br>
 Sem o elif, você só teria "sim" ou "não" (if ou else).

Com o elif, você pode ter vários caminhos: <br>
"Se for A, faça isso; senão, se for B, faça aquilo; senão, se for C...".

## 2. Como ele funciona (A Lógica)

O Python testa as condições de cima para baixo. 
Assim que ele encontra uma que seja verdadeira, ele executa<br> 
o código daquela parte e pula todo o resto do bloco. <br>
Ele não perde tempo testando as outras opções se já achou a correta.

## 3. Exemplo de Sintaxe para anotar:

```python
nota = 11

if 0 <= nota <= 10:
    if nota >= 9:
        print("Excelente você recebeu um SS") 
    elif nota >= 7: # Só será testado se a nota NÃO for >= 9
        print("Muito bem você recebeu um MS")
    elif nota >= 5: # Só será testado se a nota NÃO for >= 7
        print("Passou por pouco procure melhorar você recebeu MM")
    else: # Só será executado se NENHUMA das de cima for verdadeira
        print("Você reprovou sinto muito")
else:
    print("Nota inválida")
```

## 4. Diferença importante: $elif$ vs $vários ifs$

Vários ifs soltos: O Python vai testar todos eles, um por um, mesmo que já tenha achado um verdadeiro. Isso gasta mais processamento.

if com vários elifs: É um bloco único e inteligente. O Python para no primeiro que der certo.

### 5. O elif é uma abreviação de "else if" (que significa "senão, se").

Em programação, ele serve para verificar múltiplas condições em sequência. Imagine que o if é o seu primeiro plano, o elif são as suas outras opções caso o primeiro plano falhe, e o else é o que acontece se nenhuma das opções anteriores der certo.
