# Casos com várias condições/comparções

### Estrutura:

Quando temos várias comparações, ao invés de criar if podemos usar os operadores "and" e "or" para tratar essas condições.

Funciona assim:

if condicao_1 and condicao_2:
    vai ser executado se as 2 condições forem verdadeiras ao mesmo tempo

#### outro caso

if condicao_1 or condicao_2:
    vai ser executado se pelo menos uma das condições forem verdadeiras

## Exemplo

vamos voltar ao exemplo de cáculo de meta de vendas dos funcionários. Muitas empresas atribuem bonificação ao salário dos funcionários de acordo com o resultado do funcionário e também com o resultado da empresa como um todo.

Nesse caso, a regra funciona da seguinte forma:

    • Se o funcionário vendeu mais do que a meta de vendas e a loja bateu a meta de vendas da loja, o funcionário ganha 3% do que ele vendeu em forma de bônus<br>
    • Caso o funcionário tenha batido a meta de vendas individual dele, mas a loja não tenha batido a meta de vendas da loja como um todo, o funcionário não ganha bônus.

```python
venda_funcionario = float(input("Insira a sua meta"))
valor_total_loja = float(input("Insira qual foi o valor total faturado pela loja esse mês"))
meta_vendas_individual= 3400.00 
meta_loja = 15600.00

if venda_funcionario >= meta_vendas_individual: # Verifica se bateu a meta de vendas do funcionario
    if valor_total_loja >= meta_loja: # Verifica se a meta da loja foi batida
        bonus_funcionario = venda_funcionario * 0.03
        print (f"PARABÉNS!!! Você recebeu o bônus de {bonus_funcionario:.2f}")
    else:
        print("Você bateu a meta, mas a loja não faturou o esperado, sem bônus esse mês")
else:
    print("Você não bateu a meta individual")
```

## Outro exemplo

Agora vamos levar essa análise mais a fundo.

Nessa empresa, existe um outro caso também que garante que o funcionário ganhe um bônus, independente das vendas que ele fez naquele mês.

Todo mês os diretores da empresa fazem uma avaliação qualitativa de todos os funcionários. Nessa avaliação os diretores dão uma nota de 0 a 10
funcionário. Se a nota do funcionário for 9 ou 10, ele também ganha o bônus de 3% do valor de vendas. (os bônus não são cumulativos)

```python
nota_funcionario = int(input("Digite a nota do funcionário: "))
meta_nota = 9

if meta_nota <= nota_funcionario <= 10 or (venda_funcionario >= meta_vendas_individual and valor_total_loja >= meta_loja ):
    bonus_funcionario = venda_funcionario * 0.03
    print (f"PARABÉNS!!! Você recebeu o bônus de {bonus_funcionario:.2f}")
else:
    print("Funcionário não ganhou bônus")
```
