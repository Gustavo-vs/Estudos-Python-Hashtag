# Bloco e Identação

## Estrutura

Sempre que usamos o if ou qualquer outra estrutura no Python, devemos usr a identeação para dizer ao programa onde a estrutura começa e onde ela termina.

Isso vai ajudar muito quando tivermos mais de 1 condição ao mesmo tempo e quando tivermos várias ações para fazer dentro de um if.

### Várias ações em 1 if:

if condicao:

    alguma coisa
    outra coisa
    outra coisa mais
    outra coisa ainda mais

else:

    uma coisa
    uma coisa mais
    coisa final

### Exemplo

Vamos fazer um programa abaixo:

Digamos que você precisa criar um programa para um fundo de investimentos, conseguir avaliar o resultado de uma carteira de ações e o quanto de taxa deverá ser pago.

A regra desse fundo de investimento é:

• O fundo se compromete a entregar no minimo 5% de retorno ao ano.<br>
• Caso o fundo não consiga entregar os 5% de retorno, ele não pode cobrar taxa dos seus investidores<br>
• Caso o fundo consiga entregar mais de 5% de retorno, ele irá cobrar 2% de taxa dos seus investidores<br>
• Caso o fundo consiga mais de 20% de retorno, ele irá cobrar 4% de taxa dos seus investidores.

```python
meta = 0.05
taxa = 0
rendimento = float(input("Digite o valor do seu rendimento"))
percent_rendimento = rendimento / 100

if percent_rendimento > meta:
    if percent_rendimento > 0.20:
        taxa = 0.04
        print("A taxa foi de {}".format(taxa))
    else:
        taxa = 0.02
        print("A taxa foi de {}".format(taxa))
else:
    taxa = 0
    print("A taxa foi de {}".format(taxa))
```

## Mais de uma condicao ao mesmo tempo:

if condicao_1:<br>
    o que fazer se a condição 1 for verdadeira<br>
   
    if condicao_2:<br>
        o que fazer se a condição 1 e 2 for verdadeira<br>
    else:<br>
        o que fazer se a condição 2 for falsa (mas a condição 1 é verdadeira)<br>

else:<br>
    o que fazer quando a condição 1 for falsa
