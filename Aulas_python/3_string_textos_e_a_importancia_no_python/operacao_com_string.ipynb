# Operação com String

str -> transforma número em string<br>
in -> verifica se um texto está dentro do outro<br>
operador + -> concatena strings<br>
format e {} -> substitui textos<br>
%d -> substitui números decimais<br>

```python
faturamento = 1000
custo = 500
lucro = faturamento - custo
email = 'gustavo@gmial.com'
```

Uso do str() e do concatenar com +

```python
print('O faturamento da loja foi de: ' + str(faturamento) + ' O custo foi de: ' + str(custo) + ' O lucro foi de: ' + str(lucro))

#OBS: Forma mais complicada e trabalhosa de se fazer
```

Uso do Format

```python
print('O faturamento foi de: {}'.format(faturamento))

print('O faturamento foi de: {} o custo foi de: {} e o lucro foi de: {}'.format(faturamento, custo, lucro))

print('O faturamento foi de: {0} o custo foi de: {1} e o lucro foi de: {2} O faturamento foi de: {0} o custo foi de: {1}'.format(faturamento, custo, lucro)) 
#Neste caso é quando eu quero utilizar uma variavel que ja esteja dentro do meu format para que eu não repita ela de novo 
# la dentro eu dou posições para elas começando do número 0  e quando eu for repetir uma variavel ja existente em meu format eu coloco a posição dela dentro dos {}

#OBS: Um metodo mais facil e mais linpo do que a utilização de str() e concatenação por +
```

Uso do %s e %d

```python
print('O faturamento foi de: %d. O custo foi de: %d. O lucro foi de: %d. Envie essas informações para o email %s' % (faturamento, custo, lucro, email))

#OBS: Note que o .format() e bem parecido com o %d e %s porém ele ainda continua sendo o mais utilizado pelos desenvolvedores para manipulação de strings 
#veja que para valores numéricos utilizei %d e para valores de string %s não mude isso e siga sempre essa regra pois se não seu código acusara erro
```

Uso do in (mais exercícios práticos nas próximas aulas)

```python
print('@' in email)
print('.br' in email)
```
