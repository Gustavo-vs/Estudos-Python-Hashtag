Para pegar um texto de trás para frente: texto[indice] -> onde índice é negativo
Para pegar o pedaço de um texto use: (dois pontos). texto[:indice] ou texto[indice:] ou ainda texto[indice:indice]

```python
email = 'gustavo@gmail.com'
nome = 'Gustavo Vinicius'

print(len(email)) #imprimindo a quantidade de caracteristicas com len()
print(email[-5]) #imprimendo o indice utilzando o nome da variavel e colocando a posição da letra ou do caracte

print(email[:5]) #para no caracter 5 e não mostra o carecter 5 ex: gusta
print(email[5:]) #inicia no caracter 5 e mostra todo o restante incluindo o caracter 5 ex: vo@gmail.com
print(email[:]) #mostra tudo oque tiver na variavel ex: gustavo@gmail.com
print(email [2:9]) #mostra a patir do caracter 2 e para no caracter 9 que não é mostrado 
                   #(funciona como um ranger para mostrar de um ponto especifico até outro) ex:stavo@g
```
